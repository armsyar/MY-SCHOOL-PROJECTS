### Task 2: Do a principal component analysis (18 marks)



 
#### a)	Standardise the training dataset and perform a PCA analysis on the training data set. Explain your findings in a markdown cell (6 marks).


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_val_scaled = scaler.transform(X_val)
```

```python
X_train
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
```

```python
X_train_scaled.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)
```

```python
X_train_pca = pca.fit_transform(X_train_scaled)
```

```python
X_train_pca
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
explained_variance_ratio
```

```python
cumulative_variance = explained_variance_ratio.cumsum()
```

```python
cumulative_variance
```

The training dataset was standardized to ensure all features have a mean of 0 and a standard deviation of 1.
We can compare the result by looking at X_train and X_train_scaled.

Principal Component Analysis was applied to the standardized training dataset to reduce its diminensionality while retaining as much variance as possible.

The explained_variance_ratio shows that the first principal component (PC1) captures more than half of the total variance at 52.54%.
Followed by the other components:
PC2: 17.28%
PC3: 9.79%
PC4: 6.4%
PC5: 5.03%
PC6: 3.15%
PC7: 2.03%

The cumulative_variance shows the total variance retained by adding each principal component. The first 6 components captures almost 95% of the total variance. 


#### b)	Plot a scree plot to report variances captured by each principal component. Explain your findings in a markdown cell (4 marks).


```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Create a scree plot for explained variance ratio
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.grid()
plt.show()
```

The result clearly shows that PC1 captures the most variance in the dataset, while subsequent components show progressively lesser variance. This suggest that it may be optimal to retain just the first 3 components which explain 79.62% of the variance, capturing the majority of the dataset while reducing its dimensionality.

However, overlapping regions between the classes, particularly in the middle section, indicate areas of potential misclassification. These overlaps may negatively impact the overall model performance, as they suggest that the model might struggle to clearly distinguish between some classes in these regions.

Train the model using the training set and report the performance on the validation set and test set including accuracy rate.

```python
print("=" * 50)
print("Validation Set Evaluation")
print("=" * 50)
print(f"Accuracy: {val_accuracy:.2f}\n")
print("Classification Report:\n", val_classification_rep)
print("Confusion Matrix:\n", val_confusion_mat)

print("\n" + "=" * 50)
print("Test Set Evaluation")
print("=" * 50)
print(f"Accuracy: {test_accuracy:.2f}\n")
print("Classification Report:\n", test_classification_rep)
print("Confusion Matrix:\n", test_confusion_mat)
```

The model performed significantly better on cars compared to buses and vans, likely due to the larger number of car samples in the dataset. It has the highest recall and a balanced precision-racall tradeoff, resulting in the highest f1-score.

Vans have the lowest recall in both validation (0.30) and test sets (0.37), indicating significant difficulty in identifying them correctly.

The accuracy of 67% (validation) and 70% (test) indicates the model is moderately effective but struggles with certain classes, especially vans.

Validation Confusion Matrix Insights:

Van Misclassifications: 27 were misclassified as cars, while only 13 were correctly identified.

Test Confusion Matrix Insights:

Overlapping regions in the feature space (as noted in PCA) likely contribute to misclassifications.

#### b) Report the performance on the validation set and test set by RMSE. Explain your findings in a markdown cell (6 marks)

```python
y_val_pred_lr = lgr.predict(X_val_pca)  # Predicted values for training set
```

```python
y_test_pred_lr = lgr.predict(X_test_pca)  # Predicted values for test set
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
label_encoder.fit(y_train)
```

```python
y_val_pred_encoded= label_encoder.transform(y_val_pred_lr)
```

```python
y_test_pred_encoded = label_encoder.transform(y_test_pred_lr)
```

```python
y_val_encoded = label_encoder.transform(y_val)
```

```python
y_test_encoded = label_encoder.transform(y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse_val = np.sqrt(mean_squared_error(y_val_encoded, y_val_pred_encoded))
```

```python
rmse_test = np.sqrt(mean_squared_error(y_test_encoded, y_test_pred_encoded))
```

```python
class_names = label_encoder.classes_  # ['Bus', 'Car', 'Van']
```

```python
print("=" * 50)
print("Validation Set Evaluation")
print("=" * 50)
print(f"Validation Set RMSE: {rmse_val:.2f}\n")
print(classification_report(y_val, y_val_pred_lr, target_names=class_names))


print("=" * 50)
print("Test Set Evaluation")
print("=" * 50)
print(f"Test Set RMSE: {rmse_test:.2f}\n")
print(classification_report(y_test, y_test_pred_lr, target_names=class_names))
```

The model generalizes well, with higher accuracy and better class-wise metrics on the test set compared to the validation set. This suggests it effectively learned from the training data.

The car class consistently performs the best across both sets, with high recall and f1-scores, reflecting the model's ability to identify cars correctly and minimize false positives.

While vans were challenging on the validation set, their test set metrics (precision: 0.91, recall: 0.94) show a significant improvement. This suggests the test data aligns better with the learned patterns.

Buses continue to be the most challenging class to predict, with relatively lower precision, recall, and f1-scores, especially on the validation set. This could be due to overlapping features with other classes or fewer samples.

Weighted averages closely align with the overall accuracy, indicating balanced performance across classes, though some misclassification issues remain,
